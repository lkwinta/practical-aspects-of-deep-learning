```python
from keras import models, layers
import numpy as np
```

```python
from keras.preprocessing import  image_dataset_from_directory
 
train_dataset = image_dataset_from_directory(
    "dataset_cats_dogs/train",
    labels="inferred", 
    label_mode="binary",
    class_names=None,
    color_mode="rgb",
    batch_size=20,
    image_size=(150,150),
    shuffle=True,
)

val_dataset = image_dataset_from_directory(
    "dataset_cats_dogs/val",
    labels="inferred",
    label_mode="binary",
    class_names=None,
    color_mode="rgb",
    batch_size=20,
    image_size=(150,150),
    shuffle=True,
)

test_dataset = image_dataset_from_directory(
    "dataset_cats_dogs/test",
    labels="inferred",
    label_mode="binary",
    class_names=None,
    color_mode="rgb",
    batch_size=20,
    image_size=(150,150),
    shuffle=True,
)
```

```python
network = models.Sequential()
network.add(layers.InputLayer(shape=(150, 150, 3)))
network.add(layers.Rescaling(1./255))
network.add(layers.Conv2D(32, (3, 3), activation='relu'))
network.add(layers.MaxPooling2D((2, 2)))
network.add(layers.Conv2D(64, (3, 3), activation='relu'))
network.add(layers.MaxPooling2D((2, 2)))
network.add(layers.Conv2D(128, (3, 3), activation='relu'))
network.add(layers.MaxPooling2D((2, 2)))
network.add(layers.Conv2D(128, (3, 3), activation='relu'))
network.add(layers.MaxPooling2D((2, 2)))
network.add(layers.Flatten())
network.add(layers.Dense(512, activation='relu'))
network.add(layers.Dense(1, activation='sigmoid'))

network.summary()
```

```python
from keras.optimizers import RMSprop

network.compile(optimizer=RMSprop(learning_rate=1e-4),
                loss='binary_crossentropy',
                metrics=['accuracy'])
```

```python
result = network.fit(train_dataset, epochs=30, validation_data=val_dataset)
```

```python
from matplotlib import pyplot as plt

# plot training history
plt.plot(range(1, 31), result.history['loss'], label='loss')
plt.plot(range(1, 31), result.history['accuracy'], label='accuracy')
plt.plot(range(1, 31), result.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.legend()
```

```python
test_loss, test_acc = network.evaluate(test_dataset)
 
print('test_acc:', test_acc)
```
